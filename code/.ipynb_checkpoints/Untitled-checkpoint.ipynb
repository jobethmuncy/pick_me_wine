## Doc2Vec Bigrams
Train a each color DataFrame on `text_no_color_tokens`
This is the clean `description`, `region`, `subregion`, `grape`

```python
red_tokens = df_red['text_no_color_tokens']
white_tokens = df_white['text_no_color_tokens']
rose_tokens = df_rose['text_no_color_tokens']
sparkling_tokens = df_sparkling['text_no_color_tokens']
other_tokens = df_other['text_no_color_tokens']
```

```python
bi_red = Phrases(red_tokens, min_count=1, threshold=2)
bi_white = Phrases(white_tokens, min_count=1, threshold=2)
bi_rose = Phrases(rose_tokens, min_count=1, threshold=2)
bi_sparkling = Phrases(sparkling_tokens, min_count=1, threshold=2)
bi_other = Phrases(other_tokens, min_count=1, threshold=2)            
```

Cnythia Chaing's help with bigrams and trigrams and link to https://stackoverflow.com/questions/46441876/why-do-i-get-single-letter-vocabulary-in-gensim-word2vec

```python
red_bigram_docs = []

for doc_red in red_tokens:
    red_bigram_docs.append(bi_red[doc_red])
    
# assign tags to doc
tagged_red = [TaggedDocument(bi_red, [i]) for i, bi_red in enumerate(red_bigram_docs)]

# create model
red_bi_model = Doc2Vec(vector_size = 100, min_count = 3, epochs = 30)

# build vocab
red_bi_model.build_vocab(tagged_red)

# train model
red_bi_model.train(tagged_red, 
                   total_examples = red_bi_model.corpus_count, 
                   epochs = red_bi_model.epochs)
```

```python
white_bigram_docs = []

for doc in white_tokens:
    white_bigram_docs.append(bi_white[doc])
    
# assign tags to doc
tagged_white = [TaggedDocument(bi_white, [i]) for i, bi_white in enumerate(white_bigram_docs)]

# create model
white_bi_model = Doc2Vec(vector_size = 100, min_count = 3, epochs = 30)

# build vocab
white_bi_model.build_vocab(tagged_white)

# train model
white_bi_model.train(tagged_white, 
                   total_examples = white_bi_model.corpus_count, 
                   epochs = white_bi_model.epochs)
```

```python
rose_bigram_docs = []

for doc in rose_tokens:
    rose_bigram_docs.append(bi_rose[doc])
    
# assign tags to doc
tagged_rose = [TaggedDocument(bi_rose, [i]) for i, bi_rose in enumerate(rose_bigram_docs)]

# create model
rose_bi_model = Doc2Vec(vector_size = 100, min_count = 3, epochs = 30)

# build vocab
rose_bi_model.build_vocab(tagged_rose)

# train model
rose_bi_model.train(tagged_rose, 
                   total_examples = rose_bi_model.corpus_count, 
                   epochs = rose_bi_model.epochs)
```

```python
sparkling_bigram_docs = []

for doc in rose_tokens:
    sparkling_bigram_docs.append(bi_sparkling[doc])
    
# assign tags to doc
tagged_sparkling = [TaggedDocument(bi_sparkling, [i]) for i, bi_sparkling in enumerate(sparkling_bigram_docs)]

# create model
sparkling_bi_model = Doc2Vec(vector_size = 100, min_count = 3, epochs = 30)

# build vocab
sparkling_bi_model.build_vocab(tagged_sparkling)

# train model
sparkling_bi_model.train(tagged_sparkling, 
                   total_examples = sparkling_bi_model.corpus_count, 
                   epochs = sparkling_bi_model.epochs)
```

```python
other_bigram_docs = []

for doc in other_tokens:
    other_bigram_docs.append(bi_other[doc])
    
# assign tags to doc
tagged_other = [TaggedDocument(bi_other, [i]) for i, bi_other in enumerate(other_bigram_docs)]

# create model
other_bi_model = Doc2Vec(vector_size = 100, min_count = 3, epochs = 30)

# build vocab
other_bi_model.build_vocab(tagged_other)

# train model
other_bi_model.train(tagged_other, 
                   total_examples = other_bi_model.corpus_count, 
                   epochs = other_bi_model.epochs)
```
